```python

def distribucion_vr(D,titulo='Distribucion de Varianzas'):
    
    d=[]
    y=0
    for x in D:
        y+=x
        d.append(y)
    
    """Areas de Graficacion y visualizacion de los datos"""
    fig,ax = plt.subplots(figsize=(8,5))
    
    pasos=range(len(D))
    datos=np.array(list(map(lambda x,y:[x,y],pasos,d)))
    ax.plot(datos[:,0],datos[:,1],
            ls='--',
            color='green',
            linewidth=1,
            label='Contribucion acumulada')
    plt.title(titulo)
    plt.xlabel('Componente')
    plt.ylabel('Proporci√≥n')
    plt.grid(True)
    ax.xaxis.grid(True, which='minor')
    ax.yaxis.grid(True, which='minor')
    ax.grid(which='major', color='gray', linestyle='-')
    ax.grid(which='minor', color='grey', linestyle='--')
    ax.legend()

    return ax
```

```python
from sklearn.decomposition import PCA
pca = PCA(svd_solver='auto')

Y_pca = pca.fit_transform(fst)
varianzas = fst.var().tolist()
pca_vr=pca.explained_variance_ratio_

print(pca_vr[:10])

distribucion_vr(pca_vr,"Varianza fst")
plt.axhline(0, color="black")
plt.axvline(0, color="black")

plt.xlim(0, 10)
plt.ylim(0, 1)
pca = PCA(svd_solver='auto')

Y_pca = pca.fit_transform(fst_mod)
varianzas = fst_mod.var().tolist()
pca_vr=pca.explained_variance_ratio_

print(pca_vr[:10])

distribucion_vr(pca_vr,"Varianza fst_mod")

plt.axhline(0, color="black")
plt.axvline(0, color="black")

plt.xlim(0, 10)
plt.ylim(0, 1)
```

```python

def distribucion_vr_mult(D1,D2,titulo='Distribucion de Varianzas'):
    
    d=[]
    y=0
    for x in D1:
        y+=x
        d.append(y)
    
    """Areas de Graficacion y visualizacion de los datos"""
    fig,ax = plt.subplots(figsize=(8,5))
    
    pasos=range(len(D1))
    datos=np.array(list(map(lambda x,y:[x,y],pasos,d)))
    ax.plot(datos[:,0],datos[:,1],
            ls='--',
            color='green',
            linewidth=1,
            label='Contribucion acumulada')
    plt.title(titulo)
    plt.xlabel('Componente')
    plt.ylabel('Proporci√≥n')
    plt.grid(True)
    ax.xaxis.grid(True, which='minor')
    ax.yaxis.grid(True, which='minor')
    ax.grid(which='major', color='gray', linestyle='-')
    ax.grid(which='minor', color='grey', linestyle='--')
    ax.legend()
        
    d=[]
    y=0
    for x in D2:
        y+=x
        d.append(y)
    
    """Areas de Graficacion y visualizacion de los datos"""
    
    pasos=range(len(D2))
    datos=np.array(list(map(lambda x,y:[x,y],pasos,d)))
    ax.plot(datos[:,0],datos[:,1],
            ls='--',
            color='blue',
            linewidth=1,
            label='Contribucion acumulada modulada')
    plt.title(titulo)
    plt.xlabel('Componente')
    plt.ylabel('Proporci√≥n')
    plt.grid(True)
    ax.xaxis.grid(True, which='minor')
    ax.yaxis.grid(True, which='minor')
    ax.grid(which='major', color='gray', linestyle='-')
    ax.grid(which='minor', color='grey', linestyle='--')
    ax.legend()

    return ax
```

```python
from sklearn.decomposition import PCA
pca = PCA(svd_solver='auto')

Y_pca = pca.fit_transform(w2v)
varianzas = w2v.var().tolist()
pca_vr=pca.explained_variance_ratio_

print(pca_vr[:10])

Y_pca2 = pca.fit_transform(w2v_mod)
varianzas2 = w2v_mod.var().tolist()
pca_vr2=pca.explained_variance_ratio_

print(pca_vr2[:10])

distribucion_vr_mult(pca_vr,pca_vr2,"Varianza w2v")
plt.axhline(0, color="black")
plt.axvline(0, color="black")

plt.xlim(0, 10)
plt.ylim(0, 1)
distribucion_vr_mult(pca_vr,pca_vr2,"Varianza w2v")

```
